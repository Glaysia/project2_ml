```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

from matplotlib import pyplot as plt
import pandas as pd
from pathlib import Path

# ---------- settings ----------
SRC_CSV = '../dataset/raw/project2_0.1.0_8399.csv'
DST_DIR = '../dataset/filtered_0.1.9_4'           # output directory
# 입력 변수 (order matters for training scripts)
input_cols = [
  'w1', 'l1_leg', 'l1_top', 'l1_center', 'l2',
  'h1', 'l2_tap', 'ratio', 'Tx_turns', 'Tx_tap', 'Tx_height', 'Tx_preg',
  'Rx_turns', 'Rx_tap', 'Rx_height', 'Rx_preg', 'g1', 'g2', 'Tx_space_x',
  'Tx_space_y', 'Rx_space_x', 'Rx_space_y', 'core_N_w1', 'core_P_w1',
  'Tx_layer_space_x', 'Tx_layer_space_y', 'Rx_layer_space_x',
  'Rx_layer_space_y', 'Tx_width', 'Rx_width',
]

# 출력 변수 (필요 없는 항목·중복이 있으면 수정)
output_cols = [
  'P1_Tx_W', 'P1_Rx1_W',
  'P1_Rx2_W', 'P2_Tx_W', 'P2_Rx1_W', 'P2_Rx2_W',
  'B_mean_left_T', 'B_mean_center_T', 'B_mean_right_T',
  'B_mean_top_left_T', 'B_mean_bottom_left_T', 'B_mean_top_right_T',
  'B_mean_bottom_right_T', 'Ltx_uH', 'Lrx1_uH', 'Lrx2_uH', 'M1', 'M2',
  'k1', 'k2', 'Lmt_uH', 'Lmr1_uH', 'Lmr2_uH', 'Llt_uH', 'Llr1_uH',
  'Llr2_uH', 'Magnetizing_Current_A', 'Coreloss_W', 'totalloss_W'
]
# --------------------------------

# create destination dir once
Path(DST_DIR).mkdir(parents=True, exist_ok=True)

# load source CSV
df = pd.read_csv(SRC_CSV)
df.columns = [
  'uuid', 'computer_name', 'w1', 'l1_leg', 'l1_top', 'l1_center', 'l2',
  'h1', 'l2_tap', 'ratio', 'Tx_turns', 'Tx_tap', 'Tx_height', 'Tx_preg',
  'Rx_turns', 'Rx_tap', 'Rx_height', 'Rx_preg', 'g1', 'g2', 'Tx_space_x',
  'Tx_space_y', 'Rx_space_x', 'Rx_space_y', 'core_N_w1', 'core_P_w1',
  'Tx_layer_space_x', 'Tx_layer_space_y', 'Rx_layer_space_x',
  'Rx_layer_space_y', 'Tx_width', 'Rx_width', 'P1_Tx_W', 'P1_Rx1_W',
  'P1_Rx2_W', 'P2_Tx_W', 'P2_Rx1_W', 'P2_Rx2_W',
  'B_mean_left_T', 'B_mean_center_T', 'B_mean_right_T',
  'B_mean_top_left_T', 'B_mean_bottom_left_T', 'B_mean_top_right_T',
  'B_mean_bottom_right_T', 'Ltx_uH', 'Lrx1_uH', 'Lrx2_uH', 'M1', 'M2',
  'k1', 'k2', 'Lmt_uH', 'Lmr1_uH', 'Lmr2_uH', 'Llt_uH', 'Llr1_uH',
  'Llr2_uH', 'Magnetizing_Current_A', 'Coreloss_W'
]
df['totalloss_W'] = df['Coreloss_W'].copy()
df['totalloss_W'] += df['P1_Tx_W']
df['totalloss_W'] += df['P1_Rx1_W']
df['totalloss_W'] += df['P1_Rx2_W']
df['totalloss_W'] += df['P2_Tx_W']
df['totalloss_W'] += df['P2_Rx1_W']
df['totalloss_W'] += df['P2_Rx2_W']




# drop unwanted identifier columns
df = df.drop(columns=['uuid', 'computer_name'], errors='ignore')





def plot_hist_before_after(bf_df, af_df, col):
  """
  Compare value distributions before/after filtering with side-by-side histograms.

  Parameters
  ----------
  bf_df : pandas.DataFrame
      DataFrame BEFORE filtering (must contain `col`).
  af_df : pandas.DataFrame
      DataFrame AFTER  filtering (must contain `col`).
  col : str
      Column name to plot.
  show : bool, default True
      Call plt.show(). Set False on headless servers.
  """
  import numpy as np
  import matplotlib.pyplot as plt
  from pathlib import Path

  # decide common bins using both datasets
  combined = np.concatenate([bf_df[col].values, af_df[col].values])
  max_bins = 200  
  bins = np.linspace(combined.min(), combined.max(), max_bins + 1)
  # bins = np.histogram_bin_edges(combined, bins="auto")

  # histogram counts
  cnt_before, _ = np.histogram(bf_df[col], bins=bins)
  cnt_after, _ = np.histogram(af_df[col], bins=bins)

  # bar centers and width
  centers = (bins[:-1] + bins[1:]) / 2
  width = (bins[1] - bins[0]) * 0.4 if len(bins) > 1 else 0.4

  # plotting
  fig, ax = plt.subplots(figsize=(6, 3))
  ax.bar(centers - width / 2, cnt_before, width=width,
          alpha=0.6, label="before")
  ax.bar(centers + width / 2, cnt_after, width=width,
          alpha=0.6, label="after")
  ax.set_title(f"Distribution of {col}")
  ax.set_xlabel(col)
  ax.set_ylabel("count")
  ax.legend()
  plt.tight_layout()


  plt.show(block=False); plt.pause(0.1)
  plt.close(fig)

  return fig, ax


def get_outlier(df: pd.DataFrame, col: str, weight=1.5, show=True) -> pd.DataFrame:

  import numpy as np
  data = df[col]
  q75 = np.percentile(data, 75)
  q25 = np.percentile(data, 25)
  iqr = q75 - q25
  lower_bound = q25 - weight * iqr
  upper_bound = q75 + weight * iqr



  outliers = data[(data < lower_bound) | (data > upper_bound)]
  outlier_indices = outliers.index

  ret = df.drop(outlier_indices)
  plot_hist_before_after(df,ret, col) if show else None

  return ret

# generate one CSV per target column
for tgt in output_cols:
  cols_needed = input_cols + [tgt]
  # keep only columns actually present
  cols_existing = [c for c in cols_needed if c in df.columns]
  if tgt not in df.columns:
      print(f'Skip {tgt}: not found in data')
      continue
  out_df = df[cols_existing].copy()
  # optional: drop rows where target is NaN
  out_df = out_df.dropna(subset=[tgt])

  before_df = out_df.copy()
  out_df= after_df = get_outlier(out_df, tgt, weight=1.5, show=True)


  # build safe filename
  safe_name = tgt.replace(' ', '_').replace('[', '').replace(']', '')
  out_path = Path(DST_DIR) / f'data_{safe_name}.csv'



  out_df.to_csv(out_path, index=False)
  print(f'Wrote {out_path}, shape={out_df.shape}')
```
