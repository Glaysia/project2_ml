```python
import pandas as pd
pd.set_option('display.max_columns', None)

df = pd.read_csv('../dataset/project2_0.1.3_3719.csv', delimiter=",")

df.head()
```

```python
input_keys = [
  'w1', 'l1_leg', 'l1_top', 'l1_center', 'l2', 'h1', 'l2_tap',
  'ratio', 'Tx_turns', 'Tx_tap', 'Tx_height', 'Tx_preg', 'Rx_turns',
  'Rx_tap', 'Rx_height', 'Rx_preg', 'g1', 'g2', 'Tx_space_x',
  'Tx_space_y', 'Rx_space_x', 'Rx_space_y', 'core_N_w1', 'core_P_w1',
  'Tx_layer_space_x', 'Tx_layer_space_y', 'Rx_layer_space_x',
  'Rx_layer_space_y', 'Tx_width', 'Rx_width'
]

output_keys = [
  'Ltx_uH', 'Lrx1_uH', 'Lrx2_uH', 'M1', 'M2', 'k1', 'k2', 'Lmt_uH',
  'Lmr1_uH', 'Lmr2_uH', 'Llt_uH', 'Llr1_uH', 'Llr2_uH', 'Copperloss_Tx',
  'Copperloss_Rx_1', 'Copperloss_Rx_2', 'Coreloss_W', 'B_mean_leg_left',
  'B_mean_leg_center', 'B_mean_leg_right', 'B_mean_leg_top_left',
  'B_mean_leg_bottom_left', 'B_mean_leg_top_right',
  'B_mean_leg_bottom_right'
]
```

```python
input_df = df[input_keys]
output_df = df[output_keys]
```

```python
data:dict[str,pd.DataFrame] ={}

param = 'Llt_uH'
data[param] = pd.concat([input_df, output_df[param]], axis=1)
data[param].head()
initial_row_count = data[param].shape[0]
data[param] = data[param].dropna()
final_row_count = data[param].shape[0]
print(f"Rows before dropping NaN: {initial_row_count}, Rows after dropping NaN: {final_row_count}")
```

```python
def get_outlier(
    df:pd.DataFrame, column:str, weight:float = 1.5
) -> pd.Index:
  data = df[column]
  q1 = data.quantile(0.25)
  q3 = data.quantile(0.75)
  iqr = q3 - q1
  iqr_weighted = iqr * weight
  lower_bound = q1 - iqr_weighted
  upper_bound = q3 + iqr_weighted
  outliers_idx = data[(data < lower_bound) | (data > upper_bound)].index
  return outliers_idx

outlier_idx = get_outlier(data[param], param)
print(f"Outliers in {param}: {outlier_idx}")
print(f"Number of outliers in {param}: {len(outlier_idx)}")

filtered_data = data[param].drop(outlier_idx)
```

```python
from pycaret.regression import setup as caret_setup
from pycaret.regression.oop import RegressionExperiment
from pycaret.regression import get_config, compare_models
caret: RegressionExperiment = caret_setup(
  data=filtered_data,
  target=param,
  train_size=0.8,
  n_jobs=-1,
  use_gpu=True,
)

X_train:pd.DataFrame = get_config('X_train')
y_train:pd.DataFrame = get_config('y_train')
X_test:pd.DataFrame = get_config('X_test')     
y_test:pd.DataFrame = get_config('y_test')

best_model = compare_models(
  n_select=1,
  sort='R2',
)
```
